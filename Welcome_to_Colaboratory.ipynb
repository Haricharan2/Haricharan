## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyse and visualise data. The code cell below uses <strong>numpy</strong> to generate some random data, and uses <strong>matplotlib</strong> to visualise it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

Title:Hill and Valley Prediction with Logistic Regression

Objective:The main Objective of this project is to develop a predictive model using logistic regression that can accurately classify geographical locations as either hills or valleys based on a set of inputs given.

Data Source:YBI Foundation: https://github.com/YBI-Foundation/Dataset

## **Data Preprocessing**

## **Import Library**

```
import pandas as pd
```

```
import numpy as np
```

## **Import Data**

```
df=pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Hill%20Valley%20Dataset.csv")
```

## **Describe Data**

```
df.head()
```

```
df.info()
```

```
df.describe()
```

## **Defining Columns**

```
df.columns
```

```
print(df.columns.tolist())
```

## **Shape of the DataFrame**

```
df.shape
```

## **Unique Values of y**

```
df['Class'].value_counts()
```

```
df.groupby('Class').mean()
```

## **Define Target Variable (y) and Feature Variables (X)**

```
y=df['Class']
```

```
y.shape
```

```
y
```

```
X=df.drop('Class',axis=1)
```

```
X.shape
```

```
X
```

## **Data Visualization**

```
import matplotlib.pyplot as plt
```

```
plt.plot(X.iloc[0,:])
plt.title('Valley');
```

```
plt.plot(X.iloc[1,:])
plt.title('Hill');
```

## **Standardizing X variable**

```
from sklearn.preprocessing import StandardScaler
```

```
s=StandardScaler()
```

```
X=s.fit_transform(X)
```

```
X
```

```
X.shape
```

## **Train Test Split**

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,stratify=y,random_state=2529)
```

```
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

## **Modeling**

```
from sklearn.linear_model import LogisticRegression
```

```
Lr=LogisticRegression()
```

## **Model Evaluation**

```
Lr.fit(X_train,y_train)
```

## **Prediction**

```
y_pred=Lr.predict(X_test)
```

```
y_pred.shape
```

```
y_pred
```

```
Lr.predict_proba(X_test)
```

## **Performance Evaluation Metric**

```
from sklearn.metrics import confusion_matrix, classification_report
```

```
print(confusion_matrix(y_test,y_pred))
```

```
print(classification_report(y_test,y_pred))
```
